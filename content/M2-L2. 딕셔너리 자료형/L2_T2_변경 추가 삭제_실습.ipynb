# **Topic 2. 변경, 추가, 삭제_실습**

**1) 다음 딕셔너리 member의 '지역'을 '부산'에서 '서울'로 변경하고 확인하세요.**

[결과 예시]

~~~
{'이름': '홍길동', '지역': '서울'}
~~~

```python
member = {'이름': '홍길동',
          '지역': '부산'}

print(member)
```

**2) 다음 딕셔너리 member가 [결과 예시] 처럼 조회되게 키(key)와 값(value)를 추가하고 확인하세요.**

[결과 예시]

~~~
{'이름': '홍길동', '지역': '서울', '성별': '남자'}
~~~

```python
member = {'이름': '홍길동',
          '지역': '서울'}

print(member)
```

**3) 다음 딕셔너리 member가 [결과 예시] 처럼 조회되게 키(key)와 값(value)를 추가하고 확인하세요.**

[결과 예시]

~~~
{'이름': '홍길동', '지역': '서울', '성별': '남자', '취미': ['등산', '영화감상', '파이썬코딩']}
~~~

```python
member = {'이름': '홍길동',
          '지역': '서울',
          '성별': '남자'}

print(member)
```

**4) 다음 딕셔너리 member에서 '성별'을 삭제하고 확인하세요.**

[결과 예시]

~~~
{'이름': '홍길동', '지역': '서울', '취미': ['등산', '영화감상', '파이썬코딩']}
~~~

```python
member = {'이름': '홍길동',
          '지역': '서울',
          '성별': '남자',
          '취미': ['등산', '영화감상', '파이썬코딩']}

print(member)
```
