**7) 다음 리스트의 요소들에 대한 최댓값, 최솟값, 합계, 평균을 확인하세요.**

[결과 예시]

~~~
최댓값: 95
최솟값: 60
합계: 780
평균: 78.0
~~~

```python
score = [80, 75, 70, 95, 90, 65, 85, 70, 90, 60]
```

```python
print('최댓값:', max(score))
print('최솟값:', min(score))
print('합계:', sum(score))
print('평균:', sum(score)/len(score))
```
