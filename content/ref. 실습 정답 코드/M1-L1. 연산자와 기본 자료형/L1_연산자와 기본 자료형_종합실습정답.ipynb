**2) 다음 구문의 결과를 확인한 후 이어지는 코드 셀에 복합 대입 연산자를 사용해 다시 작성하세요.**

**3) 다음 변수들의 합계와 평균을 tot_score, avg_score 변수에 대입한 후 출력하세요.**

[결과 예시]

~~~
합계: 1104 평균: 220.8
~~~

```python
# 변수 선언
score1 = 153
score2 = 204
score3 = 259
score4 = 221
score5 = 267

# 합계 구하기
tot_score = score1 + score2 + score3 + score4 + score5

# 평균 구하기
avg_score = tot_score / 5

# 출력
print('합계:', tot_score, '평균:', avg_score)
```

**4) 홍길동의 과목별 점수는 아래와 같습니다.**

- 국어: 96, 수학: 72, 영어: 88, 역사: 80
- 변수 kor, math, eng, hist에 각 과목 점수를 대입한 뒤,
- 변수 avg에 평균 점수를 저장하여
- 문자열 '홍길동의 평균 점수는'과 avg를 하나의 print() 함수로 출력하세요.
- 힌트: 평균 = (요소 합) / (요소 개수)

[결과 예시]

~~~
홍길동의 평균 점수는 84.0
~~~

```python
# 변수 선언
kor = 96
math = 72
eng = 88
hist = 80

# 평균 구하기
avg = (kor + math + eng + hist) / 4

# 확인
print('홍길동의 평균 점수는', avg)
```

**5) 변수 name 또는 score에 적절한 형 변환을 수행해 다음 구문의 오류를 해결하세요**

- 힌트: 정수, 실수, 문자열 등으로 형 변환을 하는 함수들이 있습니다.

[결과 예시]

~~~
홍길동 점수는 950입니다.
~~~

```python
# 변수 선언
name = '홍길동'
score = 950

# 확인
print(name + ' 점수는 ' + score + '입니다.')

# TypeError: can only concatenate str (not "int") to str
```

```python
# 변수 선언
name = '홍길동'
score = 950

# 확인
print(name + ' 점수는 ' + str(score) + '입니다.')
```
