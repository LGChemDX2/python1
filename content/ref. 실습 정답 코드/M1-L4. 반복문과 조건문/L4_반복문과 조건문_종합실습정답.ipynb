**1) 리스트 score 요소를 [결과 예시]와 같이 하나씩 출력하세요.**

- 힌트: enumerate() 함수를 활용하세요.

[결과 예시]

~~~
0: 85
1: 75
2: 85
3: 95
4: 95
5: 70
6: 65
7: 95
8: 90
9: 80
~~~

```python
# 변수 선언
score = [85, 75, 85, 95, 95, 70, 65, 95, 90, 80]

# 요소 출력
for idx, val in enumerate(score):
    print(f'{idx}: {val}')
```

**2) 리스트 score 요소 중에서 90 이상인 수만 출력하세요.**

- 단, 반복문 확장 구문을 사용하지 마세요.

[결과 예시]

~~~
95
95
95
90
~~~

```python
# 변수 선언
score = [85, 75, 85, 95, 95, 70, 65, 95, 90, 80]

# 90 이상 출력
for x in score:
    if x >= 90:
        print(x)
```

**3) 리스트 score 요소 중에서 90 이상인 수만 추출해 리스트 result를 선언하고 확인하세요.**

- 단, 반복문 확장 구문을 사용하지 마세요.
- 힌트: 빈 리스트와 append() 메서드를 사용해 보세요.

[결과 예시]

~~~
[95, 95, 95, 90]
~~~

```python
# 변수 선언
score = [85, 75, 85, 95, 95, 70, 65, 95, 90, 80]

# 90 이상 추출
result = []
for x in score:
    if x >= 90:
        result.append(x)

# 확인
print(result)
```

**4) 위 3) 번 구문을 반복문 확장 구문으로 다시 작성하세요.**

[결과 예시]

~~~
[95, 95, 95, 90]
~~~

```python
# 변수 선언
score = [85, 75, 85, 95, 95, 70, 65, 95, 90, 80]

# 90 이상 추출
result = [x for x in score if x >= 90]

# 확인
print(result)
```

**5) 반복문을 사용해 [결과 예시]와 같은 모양을 출력하세요.**

- 힌트: 짝수인 경우와 홀수 인 경우 출력하는 문자를 다르게 해 보세요.

[결과 예시]

~~~
*
%%
***
%%%%
*****
~~~

```python
# 반복문
for i in range(1, 6):
    if i % 2 == 1:
        print('*' * i)
    else :
        print('%' * i)
```

**6) 반복문을 사용해 [결과 예시]와 같이 구구단 2단을 출력하세요.**

- 힌트1: 1부터 9까지 숫자를 차례대로 2에 곱해 보세요.
- 힌트2: 출력은 f-String을 사용하면 편리합니다.

[결과 예시]

~~~
2 x 1 = 2
2 x 2 = 4
2 x 3 = 6
2 x 4 = 8
2 x 5 = 10
2 x 6 = 12
2 x 7 = 14
2 x 8 = 16
2 x 9 = 18
~~~

```python
# 구구단 2단
for i in range(1, 10):
    print(f'2 x {i} = {2 * i}')
```
