**1) 0부터 5까지 1씩 증가시키면서 값을 출력하세요.**

[결과 예시]

~~~
0
1
2
3
4
5
~~~

```python
i = 0

while i < 6:
    print(i)
    i = i + 1
```

**2) 리스트 member의 요소를 하나씩 출력하세요.**

[결과 예시]

~~~
홍길동
일지매
한사랑
강우동
~~~

```python
member = ['홍길동', '일지매', '한사랑', '강우동']

i = 0
while i < len(member):
    print(member[i])
    i = i + 1
```

**3) 1부터 100까지 3의 배수만 더하고 확인하세요.**

[결과 예시]

~~~
1683
~~~

```python
total = 0
i = 1

while i <= 100:
    if i % 3 == 0:
        total = total + i
    i = i + 1
print(total)
```

**4) 변수 s에 대입된 문자열을 변수 num개 만큼 출력하세요.**

[결과 예시]

~~~
I Have a Dream!
I Have a Dream!
I Have a Dream!
I Have a Dream!
I Have a Dream!
~~~

```python
s = 'I Have a Dream!'
num = 5

while num > 0 :
    print(s)
    num = num - 1
```

**5) 무한루프를 사용해 1부터 100까지 3의 배수만 더하고 확인하세요.**

[결과 예시]

~~~
1683
~~~

```python
total = 0
i = 0

while True:
    i = i + 1
    if i > 100:
        break
    if i % 3 != 0:
        continue
    total = total + i

print(total)
```
