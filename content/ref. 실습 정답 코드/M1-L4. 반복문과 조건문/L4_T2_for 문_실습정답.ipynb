**1) 0부터 5까지 1씩 증가시켜가면서 값을 한 줄씩 출력하세요.**

[결과 예시]

1

3

**3) 1부터 100까지 모두 더하세요.**

[결과 예시]

~~~
5050
~~~

```python
total = 0

for i in range(1, 101):
    total += i
print(total)
```

**4) 1부터 100까지 홀수만 더하세요.**

- 힌트: for 문 안에 if 문을 사용해 홀수인 경우만 생각해 봅니다.

[결과 예시]

~~~
2500
~~~

```python
total = 0

for i in range(1, 101):
    if i % 2 == 1:
        total += i
print(total)
```

**5) 1부터 100까지 홀수는 모두 더하고, 짝수는 모두 빼세요.**

- 힌트: for 문 안에 if 문을 사용해 홀수와 짝수의 경우를 나누어 생각해 봅니다.

[결과 예시]

~~~
-50
~~~

```python
total = 0

for i in range(1, 101):
    if i % 2 == 1:
        total += i
    else :
        total -= i
print(total)
```
