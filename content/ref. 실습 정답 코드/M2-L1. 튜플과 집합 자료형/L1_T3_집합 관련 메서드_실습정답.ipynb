# **Topic 3. 집합 관련 메서드_실습**

**1) 다음에 선언된 집합s에 40을 요소로 추가하고 확인하세요.**

[결과 예시]

~~~
{32, 20, 21, 40, 25, 60, 13, 15}
~~~

```python
s = {13, 15, 20, 21, 25, 32, 60}
s.add(40)
print(s)
```

**2) 다음에 선언된 집합s에 40, 50, 55를 한 번에 요소로 추가하고 확인하세요.**

[결과 예시]

~~~
{13, 15, 20, 21, 25, 32, 40, 50, 55, 60}
~~~

```python
s = {13, 15, 20, 21, 25, 32, 60}
s.update([40, 50, 55])
print(s)
```

**3) 다음에 선언된 집합s에서 임의의 요소 하나를 삭제하고 확인하세요.**

- 단, 삭제된 요소가 무엇인지 반환 받아서 확인하세요.

[결과 예시] 결과가 다를 수 있음

~~~
{40, 13, 15, 50, 20, 21, 55, 25, 60}
32
~~~

```python
s = {13, 15, 20, 21, 25, 32, 40, 50, 55, 60}
deleted = s.pop()
print(s)
print(deleted)
```

**4) 다음에 선언된 집합s에서 요소 60을 삭제하고 확인하세요.**

[결과 예시]

~~~
{32, 40, 13, 15, 50, 20, 21, 55, 25}
~~~

```python
s = {13, 15, 20, 21, 25, 32, 40, 50, 55, 60}
s.remove(60)
print(s)
```

**5) 다음에 선언된 집합s에서 요소 70을 삭제하고 확인하세요.**

- 단, 삭제하려는 요소가 없어도 오류가 발생하면 안됩니다.

[결과 예시]

~~~
{32, 40, 13, 15, 50, 20, 21, 55, 25, 60}
~~~

```python
s = {13, 15, 20, 21, 25, 32, 40, 50, 55, 60}
s.discard(70)
print(s)
```

**6) 다음에 선언된 집합s에의 모든 요소를 삭제하고 확인하세요.**

[결과 예시]

~~~
set()
~~~

```python
s = {32, 40, 13, 15, 50, 20, 21, 55, 25, 60}
s.clear()
print(s)
```
