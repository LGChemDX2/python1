**3) 다음 구문을 수행해 Teacher 클래스를 선언하고 이후 실습을 진행하세요.**

```python
class Teacher:
    def __init__(self, name, subject):
        self.name = name
        self.subject = subject

    def introduce(self):
        print(f'{self.name} 선생님은 {self.subject} 과목을 가르칩니다.')
```

① Teacher 클래스의 인스턴스를 한 개 만들고 속성을 확인하세요.

- 단, teacher1 변수를 사용하고, 이름은 '홍길동', 과목은 '파이썬'으로 하세요.


[결과 예시]

~~~
홍길동
파이썬
~~~

```python
teacher1 = Teacher('홍길동', '파이썬')
print(teacher1.name)
print(teacher1.subject)
```
