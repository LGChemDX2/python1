## **Topic 1. 변수와 산술 연산자**

### **1) 변수**

## **Topic 2. print() 함수**

- Python에서 가장 많이 사용하는 함수가 print() 함수입니다.
- 단순 값, 변수의 값, 연산 결과 등을 확인하기 위해 print() 함수를 사용합니다.

```python
# 변수 선언
c = 10.5
d = 2.5
```

```python
# print() 함수를 사용하지 않은 경우 --> 값이 하나만 표시됨
c % d
c // d
```

```python
# print() 함수를 사용한 경우 #1
print(c % d)
print(c // d)
```

- print() 함수 안에 콤마로 구분해 값을 여러 개 전달할 수 있습니다

```python
# print() 함수를 사용한 경우 #2
print('c =', c)
print('d =', d)
print('c % d =', c % d)
print('c // d =', c // d)
```

- sep, end 매개변수를 사용해 print() 함수 출력 결과를 제어할 수 있습니다.

```python
# 구분자 지정
print('Hello', 'world.')
print('Hello', 'world.', sep=',')
print('Hello', 'world.', sep='')
```

```python
# 개행 여부 지정
print('Hello', 'world.')
print('Hello', 'world.')
print('Hello', 'world.', end='')
print('Hello', 'world.')
```

## **Topic 3. 비교 연산자**

- 값이나 변수 또는 연산 결과 등의 크기를 비교하는 연산자를 비교 연산자라고 합니다.
- 비교 연산자는 이후 조건문에서 많이 사용되니 잘 익혀두어야 합니다.

### **1) 비교 연산자**


- = 는 값을 대입하는 연산자이며, '같다' 비교는 == 연산자를 사용합니다.
- '같지않다' 비교는 <> 가 아닌 != 을 사용해야 함을 기억해야 합니다.
- Python은 문자의 대문자와 소문자를 구분하여 비교합니다.
- 비교 연산의 결과는 True 아니면 False 가 됩니다(true, false가 아닌, True, False 임).

#### **① 숫자 비교**

```python
# 변수 선언
a = 10
b = 20

# 크기 비교
print(a > b)
print(a == b)
print(a < b)
print(a != b)
```

#### **② 문자 비교**

```python
# 변수 선언
s1 = 'A'
s2 = 'a'

# 크기 비교
print(s1 > s2)
print(s1 == s2)
print(s1 < s2)
print(s1 != s2)
```

```python
# 비교 #1
print('abc' == 'abc')
print('abc' == 'Abc')

# 비교 #2
print('abc' != 'abc')
print('abc' != 'Abc')

# 비교 #3
print('abc' < 'Abc')
print('abc' < 'bbc')
```

```python
# ASCII 값 비교
print(ord('A'), ord('a'))
```

#### **③ 비교 연산자 활용**

- 비교 연산자는 조건에 따라 처리를 제어하는 조건문에 많이 사용합니다.
- 다음 구문은 Score 변수 값이 80 이상이면 'Good' 그렇지 않으면 'Bad'를 출력합니다.

```python
# 변수 선언
score = 85

# 조건문 예
if score >= 80:
    print('Good')
else:
    print('Bad')
```

### **2) 복합 대입 연산자**

- 연산과 대입을 동시에 수행할 수 있는 연산자를 복합 대입 연산자라고 부릅니다.
- 코드를 단순하게 해주지만, 너무 많이 사용하면 오히려 복잡해 질 수도 있으니 적절히 사용하기 바랍니다.

```python
# 정수형 변수 선언
a = 10

# 복합 대입 연산자 사용
a += 10
a -= 2
a *= 2
a /= 5

# 확인
print(a)
```

- 위 구문은 다음 구문과 같습니다.

```python
# 정수형 변수 선언
a = 10

# 복합 대입 연산자 사용 안함
a = a + 10
a = a - 2
a = a * 2
a = a / 5

# 확인
print(a)
```

## **Topic 4. 자료형 변환**

- Python은 다양한 형태의 값을 저장하고 처리할 수 있도록 여러 가지 자료형을 제공합니다.
- 상황에 따라 이러한 자료형 사이의 형 변환이 필요할 수 있습니다.

### **1) 정수형과 실수형**


#### **① 정수형**

- 음의 정수, 0, 양의 정수(자연수)를 합해 정수형이라 합니다.

```python
# 정수형 변수
num = 5

# 확인
print(type(num))
print(num)
```

#### **② 실수형**

- 소수점을 표시할 수 있는 숫자를 실수형이라고 합니다.

```python
# 실수형 변수
num = 2.5

# 확인
print(type(num))
print(num)
```

### **2) 불리언형(Boolean)**

- 불리언(또는 부울로 발음) 자료형은 논리 자료형이라고도 합니다.

#### **① True, False**

```python
# 부울형 변수
t = True
f = False

# 자료형 확인
print(type(t))
print(type(f))
```

#### **② 부울형과 논리 연산자**

- 대문자 AND, OR, NOT이 아닌 소문자 and, or, not임에 각별히 주의합니다.

```python
# 부울형과 논리 연산자
t = True
f = False

print(t and f)
print(t or f)
print(not t)
```

### **3) 자료형 사이의 변환**

- float(), int(), str(), bool() 함수를 사용해 자료형을 변환합니다.

#### **① float() 함수: 문자열, 정수 → 실수**

```python
# 변수 선언
s = '10.5'
n = 10

# 문자열, 정수 --> 실수
a = float(s)
b = float(n)

# 확인
print(type(a), a)
print(type(b), b)
```

#### **② int() 함수: 문자열, 실수 → 정수**

```python
# 변수 선언
s = '10'
n = 123.45

# 문자열, 실수 --> 정수
a = int(s)
b = int(n)

# 확인
print(type(a), a)
print(type(b), b)
```


#### **③ str() 함수: 숫자 → 문자열**

```python
# 변수 선언
n = 123.45

# 숫자 --> 문자열
s = str(n)

# 확인
print(type(s), s)
```

#### **④ bool() 함수: 숫자, 문자열 → 불리언형**

- 0이면 False, 나머지는 True로 변환됩니다.

```python
# 변수 선언
n1 = -12
n2 = 0

# 숫자 --> 불리언형
a = bool(n1)
b = bool(n2)

# 확인
print(type(a), a)
print(type(b), b)
```


- 완전 공백('')이면 False, 나머지는 True로 변환됩니다.

```python
# 변수 선언
s1 = 'a'
s2 = ''

# 문자 --> 불리언형
a = bool(s1)
b = bool(s2)

# 확인
print(type(a), a)
print(type(b), b)
```

### **4) 참고: 날짜 시간 자료형**

- Python은 날짜와 시간 관련 자료형을 제공하지 않습니다.
- 필요하다면 datetime 라이브러리를 불러와 사용해야 합니다.

```python
# 라이브러리 불러오기
from datetime import datetime

# 현재 날짜와 시간 얻기
dt = datetime.now()
```

```python
# 날짜 시간 정보 표시
print('날짜와 시간:', dt)
print('년:', dt.year)
print('월:', dt.month)
print('일:', dt.day)
print('시:', dt.hour)
print('분:', dt.minute)
print('초:', dt.second)
```
